```python
df_treatments_for_investigation = df_treatments[cols_for_investigation]
df_treatments_for_investigation
```

```python
df_failed_treatments_for_investigation = df_treatments_for_investigation[df_treatments_for_investigation["REASON"] == 2]
df_failed_treatments_for_investigation = df_failed_treatments_for_investigation.drop(labels="REASON", axis=1)
df_failed_treatments_for_investigation
```
